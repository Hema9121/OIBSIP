```python
import matplotlib.pyplot as plt
import seaborn as sns


group = df.groupby(["Region"])["EstimatedUnemploymentRate(%)"].aggregate("max").reset_index()


group = group.sort_values(by="EstimatedUnemploymentRate(%)", ascending=False)

plt.figure(figsize=(12, 8))
sns.pointplot(x='Region', y='EstimatedUnemploymentRate(%)', data=group, order=group['Region'], hue='Region')
plt.ylabel('Max rate', fontsize=12)
plt.xlabel('States', fontsize=12)
plt.title("Highest Rates of Unemployment by States", fontsize=15)
plt.xticks(rotation='vertical')
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.legend(title='Region', loc='upper left')  
plt.show()
```
